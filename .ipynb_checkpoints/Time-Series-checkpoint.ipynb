1. Visualization

The first approach taken is visualizing our time series. In this case, we can intuitively search for any trend and seasonality. Also, the plots for auto-correlation can help us determine if there is any plausible model to select.  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
coin_data = pd.read_csv("SHEET2.csv", index_col=0, parse_dates=True)
coin_rtn = pd.read_csv("RTN.csv", index_col=0, parse_dates=True)
coin_nv = pd.read_csv("NV.csv", index_col=0, parse_dates=True)
print(type(coin_data))
```

```python
coin_data['BTC'].tail()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,1,1)
coin_data['BTC'].plot()
plt.title("Bitcoin Price");
```

From the plot, we can see that the price of bitcoin went up a lot higher since Novermber. There might be a possible trend associates with it. There is no obviously seasonality and auto-correlation structure, and there is no visible turning point. 

```python
btc = np.log(coin_data['BTC'])
logReturn = btc.diff().dropna()
plt.figure(figsize=(15,10))
plt.subplot(2,1,1)
logReturn.plot()
plt.title("Bitcoin Log Return");
```

In this figure, we use continuously compounding return series to analyze the change of price and the range of fluctuations. We can see that the price doesn't vary a lot from May 2020 to Novermber 2020. However, after Novermber, the volatitity is obviously higher. Thus, a GARCH model seems reasonable to use to analyze in the later section

2. ARIMA Model

In this section, we apply ARIMA techniques to analyze the our bitcoin price. Our objective is to seek a stationary and normally distributed.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(logReturn);
plot_pacf(logReturn);
```

From the ACF and PACF plots, we can see there is no obvious auto-correlation structure found. Thus, next approach is to search for which ARIMA model fits best. Here, we introduce the Pmdarima package. It contains the auto_arima function to search for the best model based on BIC. However, to achieve that, we should first check if the data is suitable for a autoregressive model. Thus, we use lag plot to see if a linear shape exists

```python
from pandas.plotting import lag_plot

fig, axes = plt.subplots(3, 2, figsize=(8, 12))
plt.title('Btc Autocorrelation plot')

# The axis coordinates for the plots
ax_idcs = [
    (0, 0),
    (0, 1),
    (1, 0),
    (1, 1),
    (2, 0),
    (2, 1)
]

for lag, ax_coords in enumerate(ax_idcs, 1):
    ax_row, ax_col = ax_coords
    axis = axes[ax_row][ax_col]
    lag_plot(btc, lag=lag, ax=axis)
    axis.set_title(f"Lag={lag}")

plt.show()
```

As the figures show, the relationships between lags seem fairly linear, thus our next step is to determine the lag term. Here we use KPSS and ADF tests to search for the lag term. To be conservative, it's better to select the max term between two methods.

```python
from pmdarima.arima import ndiffs

kpss_diffs = ndiffs(btc, alpha=0.05, test='kpss', max_d=6)
adf_diffs = ndiffs(btc, alpha=0.05, test='adf', max_d=6)
n_diffs = max(adf_diffs, kpss_diffs)

print(f"Estimated differencing term: {n_diffs}")
# Estimated differencing term: 
```

Next step is to fit our data with lag term 1 

```python
import pmdarima as pm
```

```python
fit = pm.auto_arima(btc, n=1, information_criterion = "bic")
fit.summary()
```

The final step is to predict the btc price one step ahead. Since our data is up to date. Thus, splitting the data into train and test parts is a good start to analyze if our predict did a good job. Moreover, we will update our drained model by intaking passing days' data. 


```python
stop = int(len(btc)*4/5)
y_train = btc[:stop]
test_data = btc[stop:]
auto = pm.auto_arima(y_train, n=1, information_criterion = "bic")
```

```python
from sklearn.metrics import mean_squared_error
from pmdarima.metrics import smape

model = auto  # seeded from the model we've already fit

def forecast_one_step():
    fc, conf_int = model.predict(n_periods=1, return_conf_int=True)
    return (
        fc.tolist()[0],
        np.asarray(conf_int).tolist()[0])

forecasts = []
confidence_intervals = []

for new_ob in test_data:
    fc, conf = forecast_one_step()
    forecasts.append(fc)
    confidence_intervals.append(conf)

    # Updates the existing model with a small number of MLE steps
    model.update(new_ob)

print(f"Mean squared error: {mean_squared_error(test_data, forecasts)}")
print(f"SMAPE: {smape(test_data, forecasts)}")
# Mean squared error: 0.34238951346274243
# SMAPE: 0.9825490519101439
```

```python
fig, axes = plt.subplots(2, 1, figsize=(12, 12))

# --------------------- Actual vs. Predicted --------------------------
axes[0].plot(y_train, color='blue', label='Training Data')
axes[0].plot(test_data.index, forecasts, color='green', marker='o',
             label='Predicted Price')

axes[0].plot(test_data.index, test_data, color='red', label='Actual Price')
axes[0].set_title('Bitcoin Prices Prediction')
axes[0].set_xlabel('Dates')
axes[0].set_ylabel('Prices')

#axes[0].set_xticks(np.arange(0, 7982, 1300).tolist(), df['Date'][0:7982:1300].tolist())
axes[0].legend()


# ------------------ Predicted with confidence intervals ----------------
axes[1].plot(y_train, color='blue', label='Training Data')
axes[1].plot(test_data.index, forecasts, color='green',
             label='Predicted Price')

axes[1].set_title('Prices Predictions & Confidence Intervals')
axes[1].set_xlabel('Dates')
axes[1].set_ylabel('Prices')

conf_int = np.asarray(confidence_intervals)
axes[1].fill_between(test_data.index,
                     conf_int[:, 0], conf_int[:, 1],
                     alpha=0.9, color='orange',
                     label="Confidence Intervals")

#axes[1].set_xticks(np.arange(0, 7982, 1300).tolist(), df['Date'][0:7982:1300].tolist())
axes[1].legend()
```
