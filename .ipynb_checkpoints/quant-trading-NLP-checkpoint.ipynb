```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time 
```

```python
post_data = pd.read_excel("bitcoin_post_data.xlsx", index_col=0)
post_data.index = post_data.index.to_period("D")
post_data.head()
```

```python
post_comment = pd.read_csv("bitcoin_post_comments.csv", index_col='comment_date')
dateindex = pd.Series(post_comment.index.copy())
dateindex = dateindex.apply(lambda x : time.strptime(x, "%Y-%m-%d %H:%M:%S"))
post_comment.index = dateindex.apply(lambda x : time.strftime("%Y-%m-%d", x))
post_comment.head()
```

```python
coin_data = pd.read_excel("Coins.xlsx","Sheet2", index_col=0, parse_dates=True)
coin_rtn = pd.read_excel("Coins.xlsx","RTN", index_col=0, parse_dates=True)
coin_rtn.index = coin_rtn.index.to_period("D")
coin_nv = pd.read_excel("Coins.xlsx","NV", index_col=0, parse_dates=True)
```

```python
coin_title = coin_data.columns
plt.figure(figsize=(15,6))
for i in range(0,coin_data.shape[1]):
    coin_data[coin_title[i]].plot()
plt.legend()
```

```python
plt.figure(figsize=(15,6))
for i in range(0,coin_data.shape[1]):
    coin_nv[coin_title[i]].plot()
plt.legend()
```

```python
# Simply added scores
post_index = np.unique(post_data.index)
score = post_data["score"]
daily_score = pd.Series({score.index[0]:score[0]})
j =0 
for i in range(1,len(score)):
    if score.index[i]==post_index[j]:
        daily_score[j] += score[i]
    else:
        j += 1
        daily_score = daily_score.append(pd.Series({score.index[i]:score[i]}))
```

```python
plt.figure(figsize=(15,15))
plt.subplot(2,1,1)
coin_data['BTC'].plot()
plt.title("BTC price");
plt.subplot(2,1,2)
daily_score.plot()
plt.title("Simple sum scores");
```

```python
# Score with title sentiment label
daily_score = pd.Series({score.index[0]:score[0]})
j =0 
for i in range(1,len(score)):
    if score.index[i]==post_index[j]:
        daily_score[j] += score[i]
    else:
        j += 1
        daily_score = daily_score.append(pd.Series({score.index[i]:score[i]}))

cumu = daily_score.values[0:len(daily_score)-2]*0.33 + daily_score.values[1:len(daily_score)-1]*0.66 + daily_score.values[2:len(daily_score)]      
daily_score[2:len(daily_score)] = cumu

plt.figure(figsize=(15,10))
plt.subplot(2,1,1)
coin_data['BTC'].plot()
plt.title("BTC return");
plt.subplot(2,1,2)
daily_score.plot()
plt.title("Simple sum sentiments scores");
```

```python
# TextBlob
import re
from textblob import TextBlob

titles = post_data["title"].copy()
#Remove handlers
titles = titles.apply(lambda x:re.sub('@[^\s]+','',x))
# Remove URLS
titles = titles.apply(lambda x:re.sub(r"http\S+", "", x))
# Substituting multiple spaces with single space
titles = titles.apply(lambda x:re.sub(r'\s+', ' ', x, flags=re.I))

polar = 1e99*np.ones(len(titles))
sub = 1e99*np.ones(len(titles))
for i in range(len(titles)):
    wiki = TextBlob(titles[i])
    polar[i] = wiki.sentiment.polarity
    sub[i] = wiki.sentiment.subjectivity
```

```python
# Vader
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA

sid = SIA()
vader = titles.apply(lambda x: sid.polarity_scores(' '.join(re.findall(r'\w+',x.lower()))))
vader_pos   = vader.apply(lambda x: x['pos']) 
vader_neu   = vader.apply(lambda x: x['neu']) 
vader_neg   = vader.apply(lambda x: x['neg'])

vader = vader_pos.values+(-1)*vader_neg.values
```

```python
# Score with title sentiment label
daily_score = pd.Series({score.index[0]:score[i]*polar[0]})
j =0 
for i in range(1,len(score)):
    if score.index[i]==post_index[j]:
        daily_score[j] += score[i]*polar[i]
    else:
        j += 1
        daily_score = daily_score.append(pd.Series({score.index[i]:score[i]*polar[i]}))
        
cumu = daily_score.values[0:len(daily_score)-4]*0.2 + daily_score.values[1:len(daily_score)-3]*0.4+ daily_score.values[2:len(daily_score)-2]*0.6+ daily_score.values[3:len(daily_score)-1]*0.8+ daily_score.values[4:len(daily_score)]   
daily_score[4:len(daily_score)] = cumu
```

```python
daily_score.to_csv('daily_score.csv')
```

```python
plt.figure(figsize=(15,15))
plt.subplot(2,1,1)
coin_data['BTC'].plot()
plt.title("BTC price");
plt.subplot(2,1,2)
daily_score.plot()
plt.title("Post: Filter Scores with Sentiment");
```
